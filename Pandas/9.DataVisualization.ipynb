## Data Visualization

**The `df.plot` function**

| Plot Type (`kind='...'`) | Description | Best For | Example |
| :--- | :--- | :--- | :--- |
| **`line`** (Default) | Plots values sequentially, connecting points with lines. Excellent for showing trends over time. | **Time Series** (e.g., stock prices, sales over months) | `df.plot(kind='line')` |
| **`bar`** | Plots rectangular bars with lengths proportional to the values they represent. | **Comparing discrete categories** (e.g., sales per region, counts of items). | `df.plot(kind='bar')` |
| **`hist`** | Creates a histogram, showing the distribution of a single variable's values by grouping them into bins. | **Distribution** and frequency of continuous data (e.g., ages, scores). | `df['Age'].plot(kind='hist')` |
| **`box`** | Generates a box plot (boxplot/box-and-whisker plot) to display the distribution of data based on a five-number summary. | **Outlier detection** and comparing distributions across groups. | `df.plot(kind='box')` |
| **`area`** | Similar to a line plot, but the area beneath the line is filled. Useful for showing cumulative totals. | **Cumulative totals** or stacked data over time. | `df.plot(kind='area')` |
| **`scatter`** | Plots individual data points as markers, useful for showing the relationship between two variables. | **Correlation** between two columns (X and Y). | `df.plot(kind='scatter', x='Col_A', y='Col_B')` |
| **`pie`** | Displays data as slices of a pie, representing proportions of a whole. | **Proportions** of categorical data. | `df['Counts'].plot(kind='pie')` |

### The General Syntax of the `df.plot` function

| Parameter | Purpose | Example |
| :--- | :--- | :--- |
| **`kind`** | Specifies the type of plot (e.g., `'bar'`, `'hist'`). | `kind='bar'` |
| **`x`, `y`** | Used to explicitly define which columns map to the X and Y axes (crucial for `scatter`). | `x='Date', y='Price'` |
| **`figsize`** | Sets the size of the generated plot in inches. | `figsize=(10, 6)` |
| **`title`** | Sets a title for the plot. | `title='Quarterly Sales'` |
| **`subplots`** | If `True`, creates a separate subplot for each column. | `subplots=True` |

```python
# Import required libraries for data analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import yfinance as yf

# Set pandas display option for better float formatting
pd.set_option('display.float_format', lambda x: '%.2f' % x) #setting float format to 2 decimal places, this will help in better visualization of data

# Define the company ticker symbol
company = "AAPL"

# Define start and end dates for data download
start = dt.datetime(2020, 1, 1) # January 1, 2020 with dt package
end = dt.datetime(2025, 1, 1) # January 1, 2025 with dt package

# Download stock data using yfinance
# The function returns a DataFrame directly
stocks = yf.download(company, start=start, end=end)

# Clean column names by removing the 'Ticker' multi-index level
stocks.columns = stocks.columns.droplevel(1) # remove the 'Ticker' multi index level

# Display the first few rows of the data
stocks.head()
```

```python
# Visualize the closing price over time using a line plot (default kind)
stocks['Close'].plot(title='Closing Price Over Time') #default plot kind is line so we do not need to put it in
```

```python
# Visualize the distribution of trading volume using a histogram
stocks['Volume'].plot(kind='hist', bins=20, title='Volume Distribution') # set kind to hist with 20 bins
```

```python
# Visualize the closing price over time with a filled area plot
stocks['Close'].plot(kind='area', title='Closing Price Over Time') #same as line plot but the area below is filled in
```

```python
# Visualize all stock metrics in separate subplots
stocks.plot(title='Closing Price Over Time',  # Sets the title for the overall figure
            subplots=True,  # Creates separate subplots for each column
            layout=(2,3),  # Arranges subplots in 2 rows and 3 columns
            figsize=(12, 8),)  # Sets the figure size to 12x8 inches
```
